```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv(r'/content/loan_data.csv')
df.head(20)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# prompt: df = df["person_income ", "loan_amnt ", "credit_score", "previous_loan_defaults_on_file", "person_emp_exp", "loan_intent "])df.head()

df = df[["person_income", "loan_amnt", "credit_score", "previous_loan_defaults_on_file", "person_emp_exp", "loan_intent", "loan_status"]]
df.head()
```

```python
df['previous_loan_defaults_on_file']=df['previous_loan_defaults_on_file'].astype('category').cat.codes
df['loan_intent']=df['loan_intent'].astype('category').cat.codes
df.head()
```

```python
x= df.drop('loan_status', axis =1)
x.head()
```

```python
y = df['loan_status']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
# from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier
```

```python
#Using Logistic Regression
# lg=LogisticRegression()
# loan_ap =lg.fit(X_train, y_train)

# using Random forest
lg = RandomForestClassifier(n_estimators=100, random_state=42)
loan_ap =lg.fit(X_train, y_train)
```

```python
print('predicted: ', loan_ap.predict(X_test))
print("")
print('original: ', np.array(y_test))
```

```python
loan_ap.score(X_test, y_test)
```

```python
# prompt: use majority to solve the data imbalance

# ... (Your existing code) ...

# Check for data imbalance in the target variable 'loan_status'
y_train.value_counts(normalize=True)  # Calculate the percentage of each class
# or
y.value_counts()  # Count the number of instances in each class

# You can also visualize the class distribution using a bar plot:
y.value_counts().plot(kind='bar')
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()

# Handle data imbalance using the majority class technique
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df[df.loan_status == 0]
df_minority = df[df.loan_status == 1]

# Upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,  # sample with replacement
                                 n_samples=len(df_majority),  # match number in majority class
                                 random_state=42)  # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])

# Display new class counts
df_upsampled.loan_status.value_counts()

# Split the upsampled data into features (X) and target (y)
x_upsampled = df_upsampled.drop('loan_status', axis=1)
y_upsampled = df_upsampled['loan_status']

# Split the upsampled data into training and testing sets
X_train_upsampled, X_test_upsampled, y_train_upsampled, y_test_upsampled = train_test_split(
    x_upsampled, y_upsampled, test_size=0.2, random_state=42)

# Train your model using the upsampled data
lg_upsampled = RandomForestClassifier(n_estimators=100, random_state=42)
loan_ap_upsampled = lg_upsampled.fit(X_train_upsampled, y_train_upsampled)

# Evaluate the model on the test set
print('Upsampled Model Score: ', loan_ap_upsampled.score(X_test_upsampled, y_test_upsampled))

# ... (Rest of your code) ...
```
